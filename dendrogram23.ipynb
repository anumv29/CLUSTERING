```python
from sklearn.preprocessing import normalize
data=normalize(df)
data
```

```python
newdata=pd.DataFrame(data,columns=df.columns)
newdata
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENTROGRAM')
dend=shc.dendrogram(shc.linkage(newdata,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(newdata)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(newdata['Milk'],newdata['Grocery'],c=cluster.labels_)
```
